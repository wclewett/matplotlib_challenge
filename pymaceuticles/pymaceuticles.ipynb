```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
from scipy.stats import linregress

plt.style.use(['classic'])

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
study_meta = pd.merge(study_results, mouse_metadata, how = 'left', on = 'Mouse ID')
# Display the data table for preview
study_meta
```

```python
# Checking the number of mice.
mice_unique = list(study_meta['Mouse ID'].unique())
print(f'{len(mice_unique)} mice were tested in the study.')
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
study_meta_time = study_meta.groupby(by=['Mouse ID'])
mice_counts = pd.DataFrame(study_meta_time['Timepoint'].value_counts())
mice_dup_bool = mice_counts['Timepoint'] > 1
mice_counts[mice_dup_bool]
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
meta_mice_dup_bool = study_meta['Mouse ID'] == mice_counts[mice_dup_bool].transpose().keys()[0][0]
study_meta[meta_mice_dup_bool]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
meta_mice_singulars = study_meta['Mouse ID'] != mice_counts[mice_dup_bool].transpose().keys()[0][0]
study_meta = study_meta[meta_mice_singulars]
study_meta
```

```python
# Checking the number of mice in the clean DataFrame.
mice_unique = list(study_meta['Mouse ID'].unique())
print(f'{len(mice_unique)} mice exist the clean dataset.')
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regiment.

# This method is the most straighforward, creating multiple series and putting them all together at the end.
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
regiment_group = study_meta.groupby(by=['Drug Regimen'])
summary_stats = pd.DataFrame(regiment_group['Tumor Volume (mm3)'].count())
summary_stats = summary_stats.rename(columns = {'Tumor Volume (mm3)': 'Count'})
summary_stats['Mean'] = pd.DataFrame(regiment_group['Tumor Volume (mm3)'].mean())
summary_stats['Median'] = regiment_group['Tumor Volume (mm3)'].median()
summary_stats['Standard Deviation'] = regiment_group['Tumor Volume (mm3)'].std()
summary_stats['Variance'] = regiment_group['Tumor Volume (mm3)'].std() ** 2
summary_stats['Standard Error'] = summary_stats['Standard Deviation'] / (summary_stats['Count'] ** 1/2)
summary_stats
# This method produces everything in a single groupby function
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
avg_tumor_vol = capomulin_df.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
best_time_capomulin = best_time_df[best_time_df['Drug Regimen'] == 'Capomulin']
capomulin_best_df = pd.merge(best_time_capomulin, avg_tumor_vol, how = 'left', on = 'Mouse ID')
capomulin_best_df = capomulin_best_df.rename(columns = {'Tumor Volume (mm3)_x': 'Final Tumor Vol', 
                                                        'Tumor Volume (mm3)_y': 'Mean Tumor Vol'})
plt.scatter(capomulin_best_df['Weight (g)'], capomulin_best_df['Mean Tumor Vol'])
plt.xlabel('Weight (g)')
plt.ylabel('Mean Tumor Volume')
plt.title('Average Tumor Volume vs Weight by Mouse, Capomulin Study')
```

```python
# Calculate the correlation coefficient and linear regression model 
print("The correlation coefficient between alcohol and color intensity is",
      f"{round(st.pearsonr(capomulin_best_df['Weight (g)'],capomulin_best_df['Mean Tumor Vol'])[0],2)}")
```

```python
# for mouse weight and average tumor volume for the Capomulin regimen
x_values = capomulin_best_df['Weight (g)']
y_values = capomulin_best_df['Mean Tumor Vol']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(20, 34),fontsize=15,color="red")
plt.xlabel('Weight (g)')
plt.ylabel('Mean Tumor Vol')
plt.title('Weight(g) v. Average Tumor Volume(mm3), Linear Regression')
```
